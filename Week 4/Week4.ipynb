```python
#importing necessary Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplt
import seaborn as sns
import math
import pickle 
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
used_cars = pd.read_csv('Used_Cars.csv')                            # Read the dataset
```

```python
used_cars = used_cars.dropna()                                      # Remove any null values and display the datatypes and dataset
used_cars.describe()
print(used_cars.dtypes)
display(used_cars)
```

```python
# Covert categorical variables (State, City, Make) to numerical by using label encoder from scikit learn
le = LabelEncoder()
city_label = le.fit_transform(used_cars['City'])
state_label = le.fit_transform(used_cars['State'])
make_label = le.fit_transform(used_cars['Make'])

used_cars.drop("City", axis=1, inplace=True)
used_cars.drop("State", axis=1, inplace=True)
used_cars.drop("Make", axis=1, inplace=True)

used_cars['City'] = city_label
used_cars['State'] = state_label
used_cars['Make'] = make_label

display(used_cars)
```

```python
X = used_cars.drop(columns = ['Price'])
y = used_cars['Price']
model = LinearRegression().fit(X, y)                            # Fit linear regression model on Price and rest of the columns as predictors
```

```python
model =pickle.dump( model, open( "model.p", 'wb' ) )            # Save the model as pickle file for using in app.py
```
